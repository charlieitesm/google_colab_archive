#2. Obtener el modelo de Regresión Lineal Múltiple (RLM) del modelo
Con base a dicho modelo contestar los siguientes incisos.

## Preparación de datos

Separamos las variables dependientes de la variable independiente MEDV que es el valor medio de las casas ocupadas en 1000s de USD.

```
# Separamos las variables independientes de la dependiente MEDV que es el valor
x = mydata.loc[:, mydata.columns != "MEDV"]
print(x.shape)
print(x.head())

y = mydata[["MEDV"]]
print(y.shape)
print(y.head())
```

```
print(x)
```

```
xones = sm.add_constant(x)  # Agregamos una columna de 1s
print(xones.head())
```

## Modelo de regresion lineal multivariable

```
model_all_columns = sm.OLS(y, xones).fit()
print(model_all_columns.summary())
```

##i. Interpretar el valor del coeficiente de determinación ajustado $𝑅^2$

Con un valor $\bar{R^2}=0.734$ podemos apreciar que se tiene una correlación importante entre las variables independientes y la variable dependiente, explicándose con un 73.4% de la variabilidad de la variable dependiente queda explicada con las 13 variables independientes utilizadas.


##ii. Determina qué variables no resultaron significativas considerendo un 𝑣𝑎𝑙𝑜𝑟-𝑝 = 0.01.

En el reporte del modelo se puede apreciar que INDUS y AGE poseen valores p de 0.738 y 0.958 respectivamente, que nos indican que probablemente no están aportando mucho valor a la predicción de la variable dependiente. 


##iii. Obtener un segundo modelo de RLM omitiendo las variables que no resultaron significativas en el inciso anterior.

Comparar su coeficiente de determinación ajustado con base con respecto al primer modelo.

```
# Removemos las columnas INDUS y AGE
x2 = x.loc[:, x.columns.difference(['INDUS', 'AGE'])]
print(x2.shape)
print(x2.head())

xones2 = sm.add_constant(x2)  # Agregamos una columna de 1s

model2 = sm.OLS(y, xones2).fit()
print(model2.summary())
```

---

Después de remover las columnas INDUS y AGE y con un valor $\bar{R^2}=0.735$ podemos apreciar que hay una ganancia de 0.001 que parece no ser mucho, sin embargo es significativo ya que comprueba la suposición de que no estaban aportando mucho al modelo. Aún más interesante, los valores P de las variables restantes se ven bastante bien al acercarse a 0, lo que indica que los coeficientes son los mejores que se pueden obtener.

Este modelo tiene el potencial de ser el mejor que se puede obtener para este problema.

### Tercer modelo RLM

```
# Dejamos solo las columnas
x3 = x.loc[:, ["INDUS", "NOX", "AGE", "TAX", "LSTAT"]]
print(x3.shape)
print(x3.head())

xones3 = sm.add_constant(x3)  # Agregamos una columna de 1s

model3 = sm.OLS(y, xones3).fit()
print(model3.summary())
```

---

Se observa que el valor $\bar{R^2}=0.561$ es un tanto menor que el modelo con más variables, por el valor P podemos observar que en general los valores tienden más a 0 que indican que se están obteniendo los mejores parámetros que se pueden obtener con este set de entrenamiento.

Sin embargo, el valor P de NOX se disparó a 0.537, lo que parece indicar que dada la combinación de columnas, NOX no aporta mucho a la determinación de la variable dependiente. Pero el hecho de que NOX tuviera un valor P = 0 en el modelo con todas las variables pudiera sugerir más bien que la significancia de NOX depende de la presencia de otras variables independientes con las que tenga una fuerte correlación, como DIS.

Partiendo del supuesto anterior, podríamos sugerir un modelo con otro set de columnas que reemplace por ejemplo AGE por DIS, dado que AGE no aportaba mucho en nuestro modelo original y DIS posee una fuerte relación con NOX en la matriz de correlación. Podemos remover INDUS por la misma razón, TAX, NOX y DIS tenían fuerte relación con INDUS, así que esos pueden quedarse. Podríamos agregar RAD dada la fuerte relación con TAX.


```
x3_2 = x.loc[:, ["NOX", "DIS", "TAX", "LSTAT", "RAD"]]

model3_2 = sm.OLS(y, sm.add_constant(x3_2)).fit()
print(model3_2.summary())
```

---

Podemos así ver que hay una mejora $\bar{R^2}=0.590$, el valor P de NOX se reduce considerablemente y en general el resto de los valores P se ven mucho mejor en comparación, lo que parece aceptar nuestro supuesto sobre la significancia de las variables y que el aporte que pudieran dar INDUS y AGE están mejor codificadas en las variables con las que tienen fuerte correlación.
