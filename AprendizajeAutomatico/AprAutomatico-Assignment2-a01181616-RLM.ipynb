## Preparaci贸n de datos

Separamos las variables dependientes de la variable independiente MEDV que es el valor medio de las casas ocupadas en 1000s de USD.

```
# Separamos las variables independientes de la dependiente MEDV que es el valor
x = mydata.loc[:, mydata.columns != "MEDV"]
print(x.shape)
print(x.head())

y = mydata[["MEDV"]]
print(y.shape)
print(y.head())
```

```
print(x)
```

```
xones = sm.add_constant(x)  # Agregamos una columna de 1s
print(xones.head())
```

## Modelo de regresion lineal multivariable

```
model_all_columns = sm.OLS(y, xones).fit()
print(model_all_columns.summary())
```

##iii. Obtener un segundo modelo de RLM omitiendo las variables que no resultaron significativas en el inciso anterior.

### Tercer modelo RLM

```
# Dejamos solo las columnas
x3 = x.loc[:, ["INDUS", "NOX", "AGE", "TAX", "LSTAT"]]
print(x3.shape)
print(x3.head())

xones3 = sm.add_constant(x3)  # Agregamos una columna de 1s

model3 = sm.OLS(y, xones3).fit()
print(model3.summary())
```

Se observa que el valor $\bar{R^2}=0.561$ es un tanto menor que el modelo con m谩s variables, por el valor P podemos observar que en general los valores tienden m谩s a 0 que indican que se est谩n obteniendo los mejores par谩metros que se pueden obtener con este set de entrenamiento.

Partiendo del supuesto anterior, podr铆amos sugerir un modelo con otro set de columnas que reemplace por ejemplo AGE por DIS, dado que AGE no aportaba mucho en nuestro modelo original y DIS posee una fuerte relaci贸n con NOX en la matriz de correlaci贸n.
