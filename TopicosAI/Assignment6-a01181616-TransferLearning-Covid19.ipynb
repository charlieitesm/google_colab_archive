### Arquitecturas de CNN

Realiza un breve resumen de cad una de las siguientes arquitecturas CNN

#### 1. VGG

Propuesta por Simonyan y Zisserman su paper _Very Deep Convolutitonal Networks for Large Scale Image Recognition_ (2014).

Se caracteriza por su simplicidad, usa capas convolucionales de 3x3 puestas unas sobre otras. Usa MaxPooling para reducir su volumen y tiene 2 capas Fully-Connected de 4,096 nodos cada una seguida de un clasificador Softmax.

El número 16 y 19 en sus versiones indican el número de capas con pesos en la red. Tiene 2 grandes desventajas:

1. Es relativamente lenta para entrenar
2. La arquitectura puede ser muy grande en términos de disco y ancho de banda. 

#### 2. ResNet

A diferencia de otras arquitecturas secuenciales como AlexNet, OverFeat y VGG, ResNet forma una arquitectura basada en módulos de micro-arquitecturas, también llamadas arquitecturas red-en-red.

Fue propuesta por He et al. en su paper _Deep Residual Learning for Image Recognition_ (2015). Ha demostrado que redes muy profundas pueden ser entrenadas con SGD y una función de inicialización razonable.

Aunque es más profunda que VGG, el tamaño del modelo es sustancialmente más pequeño debido al uso de un Global Average Pooling en lugar de capas FC.

#### 3. Xception

Propuesto por François Chollet en _Xception: Deep Learning with Depthwise Separable Convolutions_, es una extensión de la arquitectura Inception que reemplaca los módulos Inception con convoluciones separables profundas.

El tamaño del modelo entrenado tiende a ser de los más pequeños.

#### 4. Inception/GoogleLeNet

Fue propuesto por Szegedy et al. en su paper _Going Deeper with Convolutions_ (2014).

Su meta es actuar como un Feature Extractor multi-nivel computando convoluciones de 1x1, 3x3 y 5x5 en el mismo módulo de la red. El producto de estos filtros se apilan y se introducen en la siguiente capa de la red.

Originalmente llamada GoogLeNet, pero ahora llamado Inception vN, con N representando un número de versión propuesto por Google.

#### 5. InceptionResNet

Inspirado por el performance de ResNet, es un híbrido entre Inception y ResNet. La principal idea es introducir conexiones residuales que combinan la salida de la convolución del módulo inception a la entrada de la siguiente capa.

#### 6. MobileNet

MobileNet es una arquitectura ligera que usa convoluciones separables profundas, lo que se traduce en que se aplique una sola convolución en cada canal de color en lugar de combinarlos todos y aplanarlos. Esto permite filtrar los canales de entrada.

Usa kernels 1x1 para combinar las salidas de las capas profundas separadas al separarlos en dos capas, una para filtrar y una para combinar, su efecto es que el cómputo y el tamaño del modelo son drásticamente reducidos.

#### 7. DenseNet

En CNN estandar, la imagen de entraba atraviesa múltiples convoluciones y se obtienen features de alto nivel. En ResNet, el mapeo de identidad es propuesto para promover la propagación de gradiente.

En DenseNet, cada capa obtiene entradas adicionales de las capas anteriores y pasa su propio mapa de features a las siguientes capas usando concatenación. Cada capa está así obteniendo el "conocimiento colectivo" de las capas anteriores.

Las ventajas de esto es que se alivia el problema de los gradientes desvanecientes, refuerza la propagación de features, promueve el reuso de features y reduce el número de parámetros.

#### 8. NASNet-Mobile

NASNet-Mobile es una arquitectura de CNN escalable que consiste en células que son optimizadas usando reinforcement learning.

Cada célula contiene unas cuantas operaciones como convoluciones separables y pooling, y es repetido múltiples veces de acuerdo a la capacidad requerida.

La versión mobile tiene 12 células con 5.3 millones de parámetros y 564 millones de multiply-accumulates (MACs).

#### 9. EfficientNet

Propuesto en el paper _EfficientNet: Rethinking Model Scaling for Convolutional Neural Networks_ (2019), introduce un método de escalamiento de modelos que usa simples pero efectivos coeficientes compuestos para escalar CNNs en una manera más estructurada.

A diferencia de otros métodos de escalamiento, aquí escala cada dimensión con un conjunto fijo de coeficientes de escala.

Combinado con AutoML, se pueden tener precisiones 10 veces mejores que con otras soluciones.

#### Sobre estrategia de Transfer Learning

En todos los casos se usó Transfer Learning con pesos de ImageNet, pero no en todos ellos se entrenaron las capas
convolucionales.

Una capa Densa de 512 con un Dropout de 0.5 y una capa de clasificación Softmax para 2 clases fue suficiente para
obtener muy buenos resultados.

Una de las limitantes más grandes es que, en aquellos modelos donde era necesario hacer fine tuning de las capas de
convolución, el entrenamiento encontraba un modelo muy bueno para datos de validación, pero después este modelo se
perdía con subsecuentes convoluciones. Sumado a esto, muchas veces el modelo comenzaba a hacer overfit a set de datos
de entrada sin que esto significara un loss de validación menor o una mejor precisión de validación.

Para atacar este problema, se usaron callbacks para implementar tanto EarlyStopping como ModelCheckpoint. El model checkpoint busca guardar el modelo con el mejor accuracy de validación encontrado durante el entrenamiento. Parar el entrenamiento no necesariamente se traduce a que el mejor modelo fue el del último epoch.

Para aquellos modelos que son muy sensibles a la actualización de pesos de las capas convolucionales, esta estrategia
permitió conseguir muy buenos resultados.

En general, para aquellos modelos que requirieron fine tuning, se congelaron aproximadamente el 20% de todas las capas del modelo base
y el entrenamiento se realizó a partir de una capa elegida inspeccionado la arquitectura de uno de ellos, cuidando elegir las capas
a partir de las cuales comienza un nuevo ciclo de convoluciones para no entorpecer el proceso de feature extraction de las capas
congeladas.

### 3. Utilidades para evaluación de modelo
Las siguientes funciones nos servirán para abstraer la generación del reporte del modelo para enfocarnos únicamente en los experimentos.

```
def train_model(model_to_fit: tf.python.keras.engine.training.Model,
                epochs: int,
                batch: int,
                train_generator: tf.keras.preprocessing.image.Iterator,
                valid_generator: tf.keras.preprocessing.image.Iterator,
                callbacks: list = None):
    
    step_size_train = train_generator.n//train_generator.batch_size
    step_size_valid = valid_generator.n//valid_generator.batch_size

    t = time.process_time()
    
    trained_model_history = model_to_fit.fit(x=train_generator,
                                            steps_per_epoch=step_size_train,
                                            validation_data=valid_generator,
                                            validation_steps=step_size_valid,
                                            epochs=epochs,
                                            callbacks=callbacks)
    
    elapsed_time = time.process_time() - t
    
    print(f"\n\n ********* Training time: {elapsed_time} s.")
    return trained_model_history

def graph_loss_accuracy(h_model,
                        epochs: int,
                        subtitle: str = ""):
    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0,epochs),h_model.history["accuracy"],label="train_acc")
    plt.plot(np.arange(0,epochs),h_model.history["val_accuracy"],label="val_acc")
    plt.title("Training and Validation Accuracy" 
              + f" ({subtitle})" if subtitle else "")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.figure()
    plt.plot(np.arange(0,epochs),h_model.history["loss"],label="train_loss") 
    plt.plot(np.arange(0,epochs),h_model.history["val_loss"],label="val_loss")
    plt.title("Training and Validation Loss" 
              + f" ({subtitle})" if subtitle else "")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.show()

# Helper functions for reporting

def print_dataset_prediction_report(predictions,
                                    dataset_generator: tf.keras.preprocessing.image.Iterator,
                                    human_readable_labels: list,
                                    dataset_title: str):
    print(f"\n\n ****** {dataset_title} Dataset Prediction Report ******")
    
    # For classification_report the one_hot_encoding test labels are needed,
    #  these are passed in the validation_tuple[1] and test_tuple[1]
    print(f"\n\nClassification Report - {dataset_title}:")
    print(classification_report(dataset_generator.labels, 
                                predictions.argmax(axis=1),
                                target_names=human_readable_labels))

    print(f"\n\nConfusion Matrix: - {dataset_title}:")
    print(confusion_matrix(dataset_generator.labels,
                            predictions.argmax(axis=1)))
    
    print("\n")

def get_model_size(model: tf.python.keras.engine.training.Model) -> int:
    """
        To get the model size, let's write the model to disk, measure it, and
        remove it afterwards.
    """
    model_filename = f"{model.name}.h5"
    
    model.save(model_filename)
    
    # Transform to MB
    size = os.path.getsize(model_filename) / 1024 / 1024
    
    os.remove(model_filename)

    return size

def print_optimizer_summary(model: tf.python.keras.engine.training.Model):
    optimizer_config = model.optimizer.get_config()
    
    print(f"Optimizer Info - {optimizer_config['name']}")

    for key in optimizer_config.keys():
        print(f"\t{key}: {optimizer_config[key]}")


def load_and_evaluate_model(path_to_model: str,
                            validation_generator: tf.keras.preprocessing.image.Iterator,
                            test_generator: tf.keras.preprocessing.image.Iterator,
                            batch: int,
                            labels: list):
    
    print(f"Loading model from disk {path_to_model}...")
    saved_model = tf.keras.models.load_model(path_to_model)
    print("Model loaded!")

    # Prepare the rewind the dataset iterators
    validation_generator.reset()
    test_generator.reset()


    # --------- VALIDATION Predictions -----------

    val_model_predictions = saved_model.predict(validation_generator,
                                                batch_size=batch,
                                                #steps=STEP_SIZE_VALID
                                                )
        
    print_dataset_prediction_report(val_model_predictions,
                                    validation_generator,
                                    labels,
                                    "Validation")


    # --------- TEST Predictions -----------

    # Use the test_tuple for predictions instead of the validation
    #  The test images are in the first position of the test_tuple
    test_model_predictions = saved_model.predict(test_generator,
                                                batch_size=batch)

    print_dataset_prediction_report(test_model_predictions,
                                    test_generator,
                                    labels,
                                    "Test")
    return saved_model

def evaluate_model(model_to_fit: tf.python.keras.engine.training.Model,
                   train_generator: tf.keras.preprocessing.image.Iterator,
                   validation_generator: tf.keras.preprocessing.image.Iterator,
                   test_generator: tf.keras.preprocessing.image.Iterator,                  
                   epochs: int,
                   batch: int,
                   labels: list,
                   callbacks: list = None,
                   title: str = ""):
    
    title = "" if not title else f"{title} - "
    graph_subtitle = f"{title}Epochs: {epochs} Batch: {batch}"

    trained_model_history = train_model(model_to_fit,
                            epochs,
                            batch,
                            train_generator,
                            validation_generator,
                            callbacks=callbacks)
    
    graph_loss_accuracy(trained_model_history,
                        epochs=len(trained_model_history.epoch), # In case we early stopped,
                        subtitle=graph_subtitle)
    
    print("\n\n ************* Model Information *************\n")
    
    model_to_fit.summary()
    
    print(f"Training Dataset Size: {train_generator.n}")
    print(f"Validation Dataset Size: {validation_generator.n}")
    print(f"Test Dataset Size: {test_generator.n}")
    print(f"Batch size: {batch}")
    print(f"Epochs: {epochs}")
    print(f"Input Image Size: {model_to_fit.layers[0].input_shape}")
    print(f"Model size: {get_model_size(model_to_fit)} MB")
    print_optimizer_summary(model_to_fit)
    
    # --------- VALIDATION Predictions -----------

    # The generator must be reset before we use it for predictions because we
    #  already consumed the iterator during the training process
    valid_generator.reset()

    val_model_predictions = model_to_fit.predict(validation_generator,
                                                    batch_size=batch,
                                                    #steps=STEP_SIZE_VALID
                                                    )
    
    print_dataset_prediction_report(val_model_predictions,
                                    validation_generator,
                                    labels,
                                    "Validation")


    # --------- TEST Predictions -----------

    # Use the test_tuple for predictions instead of the validation
    #  The test images are in the first position of the test_tuple
    test_model_predictions = model_to_fit.predict(test_generator,
                                                batch_size=batch)
    
    print_dataset_prediction_report(test_model_predictions,
                                    test_generator,
                                    labels,
                                    "Test")
```

```
# Common configurations

BASE_DATA_FOLDER = "/content/drive/My Drive/Datasets/Covid19/data/"
COVID19_MODEL_FOLDERS = "/content/drive/My Drive/Datasets/Covid19/models/"
SEED = 2020
HUMAN_READABLE_LABELS = ["Negative", "Positive"]

def get_train_val_test_data_aug_gen(preprocessing_function,
                                    target_size: tuple,
                                    batch_size: int) -> tuple:
    # Given that we have few images, we'll do better with some Data Augmentation
    train_data_augmentation = ImageDataGenerator(
            rescale=1./255, # Normalize the data to be 0-1
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            preprocessing_function=preprocessing_function)

    test_augmentation = ImageDataGenerator(rescale=1./255)

    train_generator = train_data_augmentation.flow_from_directory(
        directory=BASE_DATA_FOLDER + "train/",
        target_size=target_size,
        color_mode="rgb",
        batch_size=batch_size,
        class_mode="categorical",
        shuffle=True,
        seed=SEED,
        #save_to_dir="/content/drive/My Drive/Datasets/Covid19/augmented",
        #save_prefix="aug_"
    )

    valid_generator = test_augmentation.flow_from_directory(
        directory=BASE_DATA_FOLDER + "valid/",
        target_size=target_size,
        color_mode="rgb",
        batch_size=batch_size,
        class_mode="categorical",
        shuffle=False,
        seed=SEED
    )

    test_generator = test_augmentation.flow_from_directory(
        directory=BASE_DATA_FOLDER + "test/",
        target_size=target_size,
        color_mode="rgb",
        batch_size=batch_size,
        class_mode=None,
        shuffle=False
    )

    return train_generator, valid_generator, test_generator
```

```
efficientnet_batch = 32

train_generator, valid_generator, test_generator = get_train_val_test_data_aug_gen(
    lambda x: x, # We don't need to do further preprocessinf for EfficientNet
    (224, 224),
    efficientnet_batch)

base_model = tf.keras.applications.EfficientNetB4(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

# Freeze the weights on the feature extractor so as to keep them!
# Based on experimentation, the predictions run better by using the architecture
#  and only freezing the first layers and letting the CNN learn features of the
#  covid images
for layer in base_model.layers[:236]:
    layer.trainable = False

for layer in base_model.layers[236:]:
    layer.trainable = True

# Add an FC and Classifier layers

efficientnet_model = tf.keras.models.Sequential([
                base_model,
                
                tf.keras.layers.GlobalAveragePooling2D(),

                tf.keras.layers.Dense(512, activation='relu'),

                # Dropout
                tf.keras.layers.Dropout(0.5),

                #Output   
                tf.keras.layers.Dense(2, activation=tf.nn.softmax)
            ], name = "EfficientNet-Covid19")

# Add an optimizer

efficientnet_model.compile(optimizer=tf.keras.optimizers.RMSprop(
                                learning_rate=0.001,
                                momentum=0.9,
                            ),
                            loss='categorical_crossentropy',
                            metrics=['accuracy'])

# Callbacks to stop training if loss doesn't get better after 3 epochs
#  and to save a checkpoint of the best model seen during training
efficientnet_es = tf.keras.callbacks.EarlyStopping(
                monitor="val_loss",
                min_delta=0.01,
                patience=5,
                verbose=1,
                mode="min",
                #baseline=1.0,
                restore_best_weights=False,
            )

efficientnet_mc = tf.keras.callbacks.ModelCheckpoint(
                COVID19_MODEL_FOLDERS + "efficientnet",
                monitor="val_accuracy",
                mode="max",
                save_best_only=True,
                verbose=1,
            )

evaluate_model(efficientnet_model,
                train_generator,
                valid_generator,
                test_generator,                  
                epochs=10,
                batch=efficientnet_batch,
                labels=HUMAN_READABLE_LABELS,
                callbacks = [efficientnet_es, efficientnet_mc],
                title = "EfficientNet")
```
