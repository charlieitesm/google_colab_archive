# 1. Describe la clase de strings que representan cada una de las siguientes expresiones regulares:

1. `[a-zA-Z]+`
    * Una o más letras de la A a la Z, sin importar si son mayúsculas o minúsculas.

1. `[A-Z][a-z]*`
    * Una letra mayúscula de la A a la Z concatenado a ninguna o infinitas letras minúsculas

1. `p[aeiou]{,2}t`
    * p minúscula seguido de ninguna o a lo más 2 vocales seguido de una t minúscula

1. `\d+(\.\d+)?`
    * Al menos 1 dígito seguido opcionalmente de: un punto decimal y uno o más dígitos.

1. `([^aeiou][aeiou][^aeiou])*`
    * Cadena vacía o muchos grupos de: una consonante, seguido de una vocal, seguido de una consonante.

1. `\w+|[^\w\s]+`
    * Una cadena de al menos uno o más caracteres alfanuméricos o una cadena de uno o más caracteres que no sean alfanumérico o espacio, por ejemplo, un signo de puntuación o símbolo.

# 2. A partir del documento “austen-sense.txt” del corpus gutenberg de NLTK responde a los siguientes incisos:

1. Cuántas palabras (strings) tiene dicho documento y cuántos diferentes.
    * Austen Sense contiene: 141576 cadenas.
    * De las cuales 6833 son únicas.
1. Usa expresiones regulares para contar únicamente palabras diferentes en la novela, es decir, no contabilizar signos de puntuación o caracteres especiales.
    * Austen Sense contiene: 120787 palabras compuestas exclusivamente de caracteres alfanuméricos.
    * De las cuales 6336 son palabras únicas.
1. ¿Cuántas referencias a años contiene la novela? ¿Cuántos de estos años son diferentes?
    * Los años encontrados fueron sólo: ['1811']


```
import nltk
import re

nltk.download('gutenberg')
```

```
austen_sense = nltk.corpus.gutenberg.words('austen-sense.txt')
words = [w for w in austen_sense]
```

```
print(f"Austen Sense contiene: {len(words)} cadenas.")
print(f"De las cuales {len(set(words))} son únicas.")
```

```
# Reemplazamos todos los caracteres no alfanumericos con cadena vacia y convertimos a minúsculas
alphanumeric_tokens = [re.sub(r"[^a-zA-Z0-9]", "", w.lower()) for w in words]

# Purgamos cadenas vacias de la lista que pudieron quedar cuando se quitaron símbolos del documento
alphanumeric_tokens = [w for w in alphanumeric_tokens if w]
```

```
print(f"Austen Sense contiene: {len(alphanumeric_tokens)} palabras compuestas exclusivamente de caracteres alfanuméricos.")
print(f"De las cuales: {len(set(alphanumeric_tokens))} son palabras únicas.")
```

```
# Busquemos todos los grupos de exactamente 4 dígitos
year_tokens = [w for w in alphanumeric_tokens if re.search(r"^[0-9]{4}$", w)]
print(f"Los años encontrados fueron: {year_tokens}")
```

    * `http[s]?:\/\/[\w.]*[\w]+[/\w\?\=\&]*`. http seguido opcionalmente de una s, seguido de :// seguido opcionalmente de varias repeticiones de alfanumericos seguidos de punto, seguido de al menos un caracter alfanumerico, seguido opcionalmente de 0 o varias repeticiones de una diagonal seguida de caracteres alfanumericos, el simbolo ?, el simbolo =, el simbolo &
1. Encuentra una expresión regular para borrar del documento todos los nombres de usuario, hashtags y ligas http.
    * `@[\w]+|#[\w]+|http[s]?:\/\/[\w.]*[\w]+[/\w\?\=\&]*`. Para hacer matching de todos los nombres de usuario, hashtags y ligas http(s) y se puede reemplazar todo este match con una cadena vacía.


```
doc = ["Employers have an important role to play in supporting vital #covid19 vaccination efforts. @choo_ek, @DraCoquiMD, @DrAlethaMaybank, @drjootz and I offer thoughts on how best they can do so in @USATODAY",
       "A legend. But also a lovely, warm, utterly grounded human being. Thank you, Cicely, for the art you made, the path you forged, and the lives you changed.",
       "Today, our country is expected to mark 400,000 people confirmed lost to COVID-19. Tonight, President-elect @JoeBiden will lead us in a moment of national mourning to honor their lives and their memories. https://bideninaugural.org/covidmemorial/",
       "About 2.2 million people have been displaced within Ethiopia’s #Tigray region since fighting erupted there in Nov. with about half fleeing after their homes were burned down. How many more millions? @JoeBiden @AmbMKimani @UN #TigrayGenocide #BidenActNow",
       "En respuesta a @HillaryClinton @music4ever77 y @JoeBiden #Thanks #LoveIsInTheAir #loveandroses https://t.co/vfYsNhZTVB?amp=1 #WanderersLaSerenaxTNTSports"]

tokenized_doc = [d.split() for d in doc]
```

## 1. Hashtags y nombres de usuario

Vemos que para cada documento se extraen sus nombres de usuario y se mantiene una lista vacía para el segundo documento que no tiene ni un hashtag ni un nombre de usario.

```
hashtags_usernames = [[token for token in d if re.match(r"@[\w]+|#[\w]+", token)] for d in tokenized_doc]
hashtags_usernames
```

## 2. Todas las ligas HTTP y HTTPS

Se extraen las ligas para cada uno de los documentos y se pueden observar listas
vacías para aquellos documentos que no contienen.

```
http_links = [[token for token in d if re.match(r"http[s]?:\/\/[\w.]*[\w]+[/\w\?\=\&]*", token)] for d in tokenized_doc]
http_links
```

## 3. Clean text

```
clean_text = [" ".join([token for token in d if not re.match(r"@[\w]+|#[\w]+|http[s]?:\/\/[\w.]*[\w]+[/\w\?\=\&]*", token)]) for d in tokenized_doc]
clean_text
```
